$$f(x,y)=-[cos((x-0.1)y)]^2 -xsin(3x+y)$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    f1 = -(np.cos((x-0.1)*y))**2
    f2 = x*np.sin(3*x+y)
    return f1 - f2
```

```python
x = np.arange(-2.1, 2.1, 0.1)
y = np.arange(-2.6, 1.6, 0.1)
xx, yy = np.meshgrid(x, y)

zz = f(xx, yy)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(xx, yy, zz)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(40, 140)
plt.gcf().tight_layout()

plt.show()
```
