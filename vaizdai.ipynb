```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
```

```python
im = imread('liutukas.jpg')
plt.imshow(im)
plt.show()
```

```python
im_akis = im[200:250, 390:440]
```

```python
fig, ax = plt.subplots(1, 3, figsize=[8, 3.5])
ax[0].imshow(im)
ax[1].imshow(im_akis)
ax[2].imshow(0.2989 * im_akis[:, :, 0] + 0.5870 * im_akis[:, :, 1] + 0.1140 * im_akis[:, :, 2], cmap="gray")
plt.show()
```
